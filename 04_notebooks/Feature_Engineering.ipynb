#  Feature Engineering

```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine

#Importing Dataset From MYSQL 
engine = create_engine('mysql+pymysql://root:1234@localhost/product_analysis_db')
query = "SELECT * FROM product_analysis"
phones = pd.read_sql(query, engine)
print(f"Successfully loaded {len(phones)} rows from MySQL.")
phones.head()
```

```python
# Create a new column called Brand_Code
phones['Brand_Code'] = phones['Brand'].astype('category').cat.codes
```

```python
# Convert the date and pull out just the month number (1-12)
phones['Launch_Date'] = pd.to_datetime(phones['Launch_Date'])
phones['Launch_Month'] = phones['Launch_Date'].dt.month
```

```python
final_columns = ['Price_INR', 'RAM_GB', 'Storage_GB', 'Battery_mAh', 
                 'Profit_Margin', 'Brand_Code', 'Launch_Month']

model_ready_df = phones[final_columns]

# Save as a CSV so Phase 7 is faster to run
model_ready_df.to_csv('../01_data/model_ready_data.csv', index=False)
print("Phase 6 Complete! Dataset is ready for Machine Learning.")
```

## Feature Engineering Insights
Using Existing Features: I utilized the existing Profit_Margin column to help the model understand the relationship between cost and profit.

Brand Encoding: Converted categorical brand names into Brand_Code so the Machine Learning model can process them as numbers.

Time Analysis: Extracted Launch_Month from the date to identify seasonal trends in phone pricing.
 Model Readiness: Selected only the most relevant numeric features to ensure the model in Phase 7 is accurate and efficient.
